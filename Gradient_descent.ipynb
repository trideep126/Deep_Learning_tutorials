## **Now we'll do the same thing from scratch**

```python
import math
def sigmoid(x):
  return (1 / (1 + math.exp(-x)))
```

```python
def prediction_function(age,affordibility):
  weighted_sum = coef[0]*age + coef[1]*affordibility + intercept   # coef[0] denotes w1 , coef[1] denotes w2 and intercept denotes b
  return sigmoid(weighted_sum)                                     # pass the weighted_sum through sigmoid function
```

```python
prediction_function(0.61,1) #compare the values with predicted test values and check whether they match
```

```python
import numpy as np
def log_loss(true,predicted):
  epsilon=1e-15
  predicted_new= [max(i,epsilon) for i in predicted]
  predicted_new= [min(i,1-epsilon) for i in predicted_new]
  predicted_new= np.array(predicted_new)
  return -np.mean(true*np.log(predicted_new) + (1-true)*np.log(1-predicted_new))
```

```python
def sigmoid_numpy(x):
  return 1/(1+np.exp(-x))
```

```python
def gradient_descent(age,affordibility,true,epochs,loss_threshold):
  w1 = w2 = 1
  bias=0
  learning_rate= 0.01
  n = len(age)

  for i in range(epochs):
    weighted_sum = w1*age + w2*affordibility + bias
    predicted = sigmoid_numpy(weighted_sum)

    loss= log_loss(true,predicted)

    w1_derivative = (1/n)*(np.dot(np.transpose(age),(predicted-true)))
    w2_derivative = (1/n)*(np.dot(np.transpose(affordibility),(predicted-true)))

    bias_derivative = np.mean(predicted-true)

    w1 = w1 - learning_rate*w1_derivative
    w2 = w2 - learning_rate*w2_derivative
    bias = bias - learning_rate*bias_derivative

    print(f"Epoch:{i},w1:{w1},w2:{w2},bias:{bias},loss:{loss}")

    if loss<=loss_threshold:
       break

  return w1,w2,bias
```
