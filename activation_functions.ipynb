```python
import math
def sigmoid(x):
  return ( 1 / (1 + math.exp(-x)))
```

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x) / (math.exp(x) + math.exp(-x)))
```

```python
def relu(x):
  return max(0,x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define some common activation functions
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def tanh(x):
  return np.tanh(x)

def relu(x):
  return np.maximum(0, x)

# Plot the activation functions
x = np.linspace(-5, 5, 100)

plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, relu(x), label='ReLU')

plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.title('Activation Functions')
plt.grid(True)
plt.legend()

# Highlight key properties
plt.axvline(x=0, color='gray', linestyle='--', label='Non-differentiable point (ReLU)')
plt.ylim(-1.2, 1.2)  # Adjust limits to show non-squashing behavior

plt.text(4, 1.1, 'Non-squashing', ha='center', va='bottom', fontsize=12)
plt.text(-1.2, 0.5, 'Centered Mean (Optional)', ha='left', va='center', fontsize=12)

plt.show()
```
