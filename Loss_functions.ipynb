**Mean Absolute Error**

```python
'''Using normal function'''
def mae(true,predicted):
  total_error = 0
  for t,p in zip(true,predicted):
    total_error += abs(t-p)
    mae= total_error/len(true)
  return print(f"MAE:",mae)
```

**Mean Squared Error**

```python
'''Using normal function'''
def mse(true,predicted):
  total_error=0
  for t,p in zip(true,predicted):
    total_error += (t-p)**2
    mse= total_error/ len(true)
  return print(f"MSE",mse)
```

```python
'''Using Numpy'''
np.mean((predicted-true)**2)
```

**Log-loss / Binary Crossentropy**

```python
def log_loss(true,predicted):
  epsilon= 1e-15
  predicted_new= [max(i,epsilon) for i in predicted]
  predicted_new= [min(i,1-epsilon) for i in predicted_new]
  predicted_new= np.array(predicted_new)
  return -np.mean(true*np.log(predicted_new)+(1-true)*np.log(1-predicted_new))
```
